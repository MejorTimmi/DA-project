# 1. Write a loop statement that prints out numbers between 1 and 20

```python
number = 2
while 1 < number < 20:
    print(number)
    number += 1
```

```python
number = range(2, 20)
for i in number:
    print(i)
        
```

# 2. Loop that prints out even number between 30 and 70

```python
num = 32
while 30 < num < 70:
    print(num)
    num += 2
```

```python
 nums = range(32, 70, 2)
for i in nums:
    print(i)
```

# 3. Loop that prints multiplication table 5

```python
print("Multiplication Table 5")
nums = range(1, 13)
for i in nums:
    print(f"5*{i} = {5*i}")
```

```python
print("Multiplication Table 5")
nums = 1
while nums >= 1 and nums <=12:
    print(f"5*{nums} = {5*nums}")
    nums += 1
```

# 4. A program that accept number input from user and return the factorial of the number.

```python
input_number = int(input("Enter a whiole number: "))
factorial = 1
if input_number < 0:
    print("There is no factorial for a negetive number")
elif input_number == 0:
    print("Factorial for 0 is 1")
else:
    for num in range(1, input_number + 1):
        factorial = factorial * num
    print(f"The factorial of {num} is {factorial}")
```

```python
input_num = int(input("Enter a whole number: "))
factorial = 1
factorial = factorial * input_num
while input_num > 0:
    print(f"The factorial of {input_num} is {factorial}")
```
